```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
rectangle_size = np.linspace(1,10,10)

lin_reward = 100-rectangle_size
exp_reward = np.exp(-rectangle_size/10)
```

```python
sigmas = np.linspace(1,10,10)

p_correct =[]
for sig in sigmas:
    p_correct.append(norm.cdf(rectangle_size, 0, sig) -  norm.cdf(-rectangle_size, 0, sig))
        
```

```python
for i_sig in range(1,len(sigmas)):
    plt.plot(rectangle_size,p_correct[i_sig])
```

```python
for i_sig in range(1,len(sigmas)):
    plt.plot(rectangle_size,lin_reward*p_correct[i_sig])
```

```python
for i_sig in range(1,len(sigmas)):
    plt.plot(rectangle_size,exp_reward*p_correct[i_sig])
```
