```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

subID = '1-NJ'

block = ['S1-NJ-B1-Number-20191024-125142.pkl',
'S1-NJ-B2-Space-20191024-130507.pkl',
'S1-NJ-B3-Number-20191024-132003.pkl',
'S1-NJ-B4-Space-20191024-132755.pkl']

# Create dictionary
data = {'Num_block':[],
        'Block_ID':[],
        'Trial_ID':[],
        'Resp_loc':[], 
        'Resp_conf':[], 
        'Correct':[],
        'RT_resp':[], 
        'RT_conf':[],
        'Wager':[],
        'X':[],
        'N':[],
        'x_mean':[],
        'Mouse_track':[]}

# Create dataframe
concat = pd.DataFrame(data)

for iblock in range(0,4):
    currblock = pd.read_pickle(block[iblock])
    concat = concat.append(currblock,ignore_index=True,sort=False)

filename_pickle = str('S' + subID + '.pkl')
concat.to_pickle(filename_pickle)

filename_csv = str('S' + subID + '.csv')
concat.to_csv(filename_csv)

#concat = pd.read_pickle('S1-NJ.pkl')
#len(concat)
```
